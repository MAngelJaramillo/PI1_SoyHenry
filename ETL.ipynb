```python
#Se importan las librerias necesarias para este proceso de ETL
import pandas as pd
import numpy as np
import re #Modulo para operaciones con Expresiones Regulares, será usado principalmente para el desanidado
```

```python
#Carga del dataset de movies
movies_df = pd.read_csv('C:/Users/march/Downloads/Movies/movies_dataset.csv', sep=',')
```

```python
#Verificación del contenido
movies_df.head()
```

```python
#Revisión a detalle
movies_df.info()
```

```python
#Acá se define la función para desanidar las columnas anidadas
#Al observar que están anidadas en diccionarios se procede a extraer los datos de la clave "name" que se encuentra en cada columna anidada
def extract_name(value):
    try:
        pattern = r"'name':\s+'(.*?)'"
        match = re.findall(pattern, value)
        return ', '.join(match)
    except (TypeError, AttributeError):
        return ''
```

```python
#Extracción de las colecciones
movies_df['collection'] = movies_df['belongs_to_collection'].apply(extract_name)
```

```python
#Revisión de las colecciones
movies_df['collection'].head(5)
```

```python
#Extracción de los géneros
movies_df['genre'] = movies_df['genres'].apply(extract_name)
```

```python
#Revisión de los géneros
movies_df['genre'].head(5)
```

```python
#Extracción de las compañias
movies_df['production_company'] = movies_df['production_companies'].apply(extract_name)
```

```python
#Revisión de las compañias 
movies_df['production_company'].head(5)
```

```python
#Extracción de los países de producción
movies_df['production_country'] = movies_df['production_countries'].apply(extract_name)
```

```python
#Revisión de los países
movies_df['production_country'].head(5)
```

```python
#Extracción de los lenguajes hablados
movies_df['language'] = movies_df['spoken_languages'].apply(extract_name)
```

```python
#Revisión de los lenguajes
movies_df['language'].head(5)
```

```python
#Ya extraidos todos los valores "name" en las columnas anidadas podemos revisar que fueron extraidos sin ningún problema
movies_df.head()
```

```python
#Se reemplazaran los valores nulos de la columna de presupuesto con 0
movies_df['budget'].fillna(0, inplace=True)
print('Valores NULOS de la columna de presupuesto:', movies_df.budget.isna().sum())
```

```python
#Se reemplazaran los valores nulos de la columna de recaudación con 0
movies_df['revenue'].fillna(0, inplace=True)
print('Valores NULOS de la columna de recaudación:', movies_df.revenue.isna().sum())
```

```python
#Se remueven los valores nulos de la columna "release_date"
movies_df['release_date'] = pd.to_datetime(movies_df['release_date'], errors='coerce')

movies_df.dropna(subset = ['release_date'], inplace=True)
```

```python
#Se verifica que los valores nulos de la columna "release_date" fueron completamente removidos
movies_df['release_date'].isnull().sum()
```

```python
#Se le dara el formato de "AAAA-mm-dd" a la columna de release_date
movies_df['release_date'] = pd.to_datetime(movies_df['release_date']).dt.strftime('%Y-%m-%d')
```

```python
#Se creará una nueva columna de nombre "release_year" que contenga el año de salida de la pelicula, esto extraido de la anterior columna
movies_df['release_year'] = pd.to_datetime(movies_df['release_date']).dt.year
```

```python
#Comprobación de ambas columnas
movies_df[['release_date','release_year']].tail()
```

```python
#Hay que crear una columna "return" con el retorno de la inversión de la pelicula, esto siendo el resultado de dividir "revenue"/"budget"
movies_df['budget'] = pd.to_numeric(movies_df['budget'], errors='coerce')
movies_df['revenue'] = pd.to_numeric(movies_df['revenue'], errors='coerce')
```

```python
#Creación y calculo de la columna "return"
movies_df['return'] = movies_df['revenue']/movies_df['budget']
```

```python
#Para evitar valores no númericos en la columna "return" se les reemplazará con 0
movies_df['return'] = pd.to_numeric(movies_df['return'], errors='coerce').fillna(0)

#Para evitar valores infinitos ya que pueden llegar a haber divisiones por 0 en las columnas de recaudación y presupuesto también reemplazaremos esos infinitos por un 0
movies_df['return'].replace([np.inf, -np.inf], 0, inplace=True)
```

```python
#Se eliminan las columnas irrelevantes
movies_df.drop(columns=['video', 'imdb_id', 'adult', 'original_title', 'poster_path', 'homepage',], inplace = True)
```

```python
#Comprobación de la eliminación y de las modificaciones hechas hasta ahora
movies_df.head(5)
```

```python
#Por último extraeré las id's de las columnas anidadas ya que pueden ser necesarias para desarrollo posterior de este MVP
#Se crea una función similar pero está busca valores númericos ya que las id's de estas columnas funcionan por integers
def extract_id(value):
    try:
        pattern = r"'id':\s*(\d+)"
        match = re.findall(pattern, value)
        return ', '.join(match)
    except (TypeError, AttributeError):
        return ''
```

```python
#Se extraen las ID's
movies_df['collection_id'] = movies_df['belongs_to_collection'].apply(extract_id)
movies_df['genre_id'] = movies_df['genres'].apply(extract_id)
movies_df['company_id'] = movies_df['production_companies'].apply(extract_id)
```

```python
#Por último se eliminarán las columnas anidadas originales ya que ya extrajimos sus datos útiles, los códigos iso no se extraeran ya que son redundantes
movies_df.drop(columns=['belongs_to_collection', 'genres', 'production_companies', 'production_countries', 'spoken_languages',], inplace = True)
```

```python
#Por último se comprueba que todas las columnas han sido desanidadas y que las transformaciones requeridas fueron exitosas
movies_df.head(5)
```

```python
#Se exporta el CSV limpio
movies_limpio = movies_df.to_csv('C:/Users/march/Downloads/Movies/movies_limpio.csv', index=False, header=True)
```
